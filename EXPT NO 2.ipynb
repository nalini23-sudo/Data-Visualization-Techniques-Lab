```python
import pandas as pd
import sqlite3

# Read Excel file
df = pd.read_excel(r"C:\Users\babyn\Downloads\archive\research_student (1).xlsx")

# Connect to SQLite database
conn = sqlite3.connect("research_student.db")

# Export data to SQLite table
df.to_sql("research_student", conn, if_exists="replace", index=False)

print("Data exported to SQLite successfully!")

conn.close()
```

```python
import pandas as pd
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect("research_student.db")

# Read data from SQLite table
df = pd.read_sql("SELECT * FROM research_student", conn)

# Display first 5 records
print("First 5 Records")
print(df.head())

# Dataset Information
print("\nDataset Information")
df.info()

# Statistical Summary
print("\nStatistical Summary")
print(df.describe(include='all'))

conn.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_excel(r"C:\Users\babyn\Downloads\archive\research_student (1).xlsx")
print(df.columns)
```

```python
branch = df["Branch"].value_counts()

plt.bar(branch.index, branch.values)

plt.title("Branch-wise Student Count")
plt.xlabel("Branch")
plt.ylabel("Number of Students")

plt.show()
```
